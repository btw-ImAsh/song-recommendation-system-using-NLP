```python
import nltk
import re
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
df_cat = df_cat.reset_index()
```

```python
df_cat['track_artist_1'] = df_cat['track_artist'].apply(lambda x: x.replace(" ",""))
df_cat['track_artist_1'] = df_cat['track_artist_1'].apply(lambda x: x.split(" "))
```

```python
df_cat['track_album_name'] = df_cat['track_album_name'].apply(lambda x: x.replace(" ",""))
df_cat['track_album_name'] = df_cat['track_album_name'].apply(lambda x: x.split())
```

```python
df_cat['playlist_subgenre'] = df_cat['playlist_subgenre'].apply(lambda x: x.replace(" ",""))
df_cat['playlist_subgenre'] = df_cat['playlist_subgenre'].apply(lambda x: x.split())
```

```python
df_cat['playlist_name'] = df_cat['playlist_name'].apply(lambda x: x.replace(" ",""))
df_cat['playlist_name'] = df_cat['playlist_name'].apply(lambda x: x.split())
```

```python
df_cat['playlist_genre'] = df_cat['playlist_genre'].apply(lambda x: x.replace(" ",""))
df_cat['playlist_genre'] = df_cat['playlist_genre'].apply(lambda x: x.split())
```

```python
df_cat['language'] = df_cat['language'].apply(lambda x: [x])
df_cat['year'] = df_cat['year'].apply(lambda x: [x])
```

```python
df_cat['tags'] = df_cat['track_album_name'] + df_cat['playlist_name'] + df_cat['playlist_genre'] + df_cat['playlist_subgenre'] + df_cat['year'] + df_cat['language'] + df_cat['track_artist_1']
```

```python
df_cat['tags'] = df_cat['tags'].apply(lambda x: " ".join(x))
```

```python
df_cat['tags']
```

```python
def stem(sent):
    sent = re.sub('[^a-zA-Z0-9]', ' ', sent)
    sent = sent.lower()
    words = nltk.word_tokenize(sent)
    updated_words = [ps.stem(word) for word in words]
    return ' '.join(updated_words)
```

```python
df_cat['tags'] = df_cat['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
df_cat = df_cat[['track_id', 'track_name', 'track_artist', 'tags']]
```

```python
df_num
```

```python
tfidf_matrix = tfidf.fit_transform(df_cat['tags'])
```

```python
from scipy.sparse import hstack
from sklearn.metrics.pairwise import cosine_similarity
```

```python
combined_matrix = hstack([tfidf_matrix, df_num])
```

```python
combined_matrix
```

```python
from scipy.sparse import coo_matrix

# If you have a coo_matrix (e.g., from TF-IDF + hstack), convert to CSR
combined_matrix = combined_matrix.tocsr()
```

```python
def get_similar_songs(song, top_n=8):
    song_index = df_cat[df_cat['track_name']==song].index[0]
    vec = combined_matrix[song_index]
    sim_scores = cosine_similarity(vec, combined_matrix).flatten()
    top_indices = sim_scores.argsort()[::-1][1:top_n + 1]
    recommended_songs_id=[]
    recommended_songs=[]
    for i in top_indices:
        recommended_songs_id.append(df_cat.iloc[i].track_id)
        recommended_songs.append(df_cat.iloc[i].track_name)
    return recommended_songs_id, recommended_songs
    
get_similar_songs("I Don't Care (with Justin Bieber) - Loud Luxury Remix")
```
